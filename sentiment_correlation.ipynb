```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('processed_df.csv')
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
stemmer = PorterStemmer()
```

```python
word_tokenize(stemmer.stem('satisfied')+' '+ stemmer.stem('quality'))
```

```python
indicators = set(word_tokenize(stemmer.stem('satisfied')+' '+ stemmer.stem('quality'))
)
```

```python
import pdb
```

```python
indicators = {}
```

```python
indicators = {}
indicator_words = []
for row in df.iterrows():
    row = row[1]
    for word in row['Review'].split():
        if word in indicators:
            indicators[word].append(row['sentiment_score'])
        else:
            indicators[word] = []

for key, value in indicators.items():
    if abs(np.mean(indicators[key])) >= 0.75:
        indicator_words.append(key)
```

```python
indicator_words
```

```python
for word in indicator_words:
    df[f'{word}_prescence'] = df['Review'].apply(lambda x: 1 if word in x else 0)
```

```python
binary_features = list(filter(lambda x: x.endswith('prescence'), df.columns))
```

```python
binary_features.append('sentiment_score')
```

```python
df[binary_features].corr(method='pearson').to_csv('correlation_analysis.csv')
```
